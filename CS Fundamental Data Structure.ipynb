Tedapat Tiga metode (method) atau fungsi yang bisa digunakan untuk menambahkan isi atau item ke List:

prepend(item) menambahkan item dari depan;
append(item) menambahkan item dari belakang.

```python
buah = ["jeruk", "apel", "mangga", "duren"]
```

```python
#list 
buah = ["jeruk", "apel", "mangga", "duren"]
# Tambahkan manggis
buah.append("manggis")
buah.append("salak")
buah
```

```python
#list mula-mula
from collections import deque

buah = ["jeruk", "apel", "mangga", "duren"]
buah = deque(buah)
buah.appendleft("anggur") 
buah
```

```python
#list mula-mula
buah = ["jeruk", "apel", "mangga", "duren"]
buah.insert(2, "duren")
buah
```

## Connect python to database PostgreSQL.

pada library psycopg2 terdapat beberapa fungsi di antaranya :

- Untuk membuka koneksi ke database, psycopg membutuhkan minimal dua argumen : nama database (‘dbname’) dan nama pengguna (‘user’), namun biasanya terdapat argumen lain yang dibutuhkan seperti (‘host’), (‘port’), dan (‘password’). Sintaks untuk membuka koneksi sebagai berikut :

`req = psycopg.connect(user=’<user>’, password=’<password>’ host=’<host>’, port=’<port>’, dbname=’<dbname>’)`

- Kemudian terdapat fungsi kursor, dimana fungsi ini berguna untuk membuka jalannya pemrosesan lanjutan dalam koneksi sebuah database, dimana fasilitas ini nantinya berguna untuk pengambilan, penambahan, dan penghapusan catatan dalam koneksi database. Fungsi ini mirip dengan konsep bahasa pemrograman iterator.
Selain itu kursor juga berfungsi sebagai identifikasi sebuah baris dalam sebuah tabel untuk di update atau di delete. Fungsi cursor dapat dijalankan dengan perintah :

`cur = conn.cursor()`

- Kemudian kita dapat melakukan eksekusi query dengan cara:

`conn.execute(‘sql query’)`

- Kemudian kita perlu menyimpan result-set hasil eksekusi query kedalam variabel, contoh variabel record :

`record = curr.fetchall()`

Kelebihan dari psycopg adalah result set hasil query secara otomatis dikonversi ke dalam variabel tipe list.


```python
import psycopg2
import pandas as pd

params_dic = {
    "host" : "localhost",
    "database" : "belajar",
    "user" : "postgres",
    "password" : "postgres",
    "port" : 5432
}

def get_data_table(params_dic, query, column_name):
    # generate function to request data from db postgres
    try :
        conn = psycopg2.connect(**params_dic)
        cursor = conn.cursor()
        cursor.execute(query)
        print("select rows from table_name using cursor.fetchall")
        table_records = cursor.fetchall()
        df = pd.DataFrame(table_records, columns=column_name)
        return df

    except (Exception, psycopg2.Error) as error:
        print("Error while fetching data from PostgreSQL", error)

    finally:
        # closing database connection.
        if conn:
            cursor.close()
            conn.close()
            print("PostgreSQL connection is closed")
```

```python
result = get_data_table(params_dic,"select username,email from accounts",['username','email'])
```
